```python
import pandas as pd
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
```

```python
df=pd.read_csv("/content/used_cars_data.csv")
df.head()
```

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
df['Location']= label_encoder.fit_transform(df['Location'])
df['Location'].unique()

df['Name']= label_encoder.fit_transform(df['Name'])
df['Name'].unique()

df['Year']= label_encoder.fit_transform(df['Year'])
df['Year'].unique()

df['Kilometers_Driven']= label_encoder.fit_transform(df['Kilometers_Driven'])
df['Kilometers_Driven'].unique()

df['Fuel_Type']= label_encoder.fit_transform(df['Fuel_Type'])
df['Fuel_Type'].unique()

df['Transmission']= label_encoder.fit_transform(df['Transmission'])
df['Transmission'].unique()

df['Owner_Type']= label_encoder.fit_transform(df['Owner_Type'])
df['Owner_Type'].unique()

df['Mileage']= label_encoder.fit_transform(df['Mileage'])
df['Mileage'].unique()

df['Engine']= label_encoder.fit_transform(df['Engine'])
df['Engine'].unique()

df['Power']= label_encoder.fit_transform(df['Power'])
df['Power'].unique()

df['Seats']= label_encoder.fit_transform(df['Seats'])
df['Seats'].unique()

df['Price']= label_encoder.fit_transform(df['Price'])
df['Price'].unique()

df.dropna(axis=1,inplace=True)
```

```python
X = df[['Name','Location','Year','Kilometers_Driven','Fuel_Type','Transmission','Owner_Type','Mileage','Engine','Power','Seats']]
Y = df[['Price']]
```

```python
X

Y

x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size =0.2,random_state=42)

x_train

x_test

y_train

y_test
```

```python
from sklearn.linear_model import LinearRegression
m1=LinearRegression()

m1.fit(x_train,y_train)

m1.predict([[700,10,2020,700000,2,2,4,200,19,17,2]])

predict=m1.predict(x_test)
predict
```

```python
from sklearn import metrics
import numpy as np                   

metrics.mean_absolute_error(y_test,predict)

np.sqrt(metrics.mean_squared_error(y_test,predict))

metrics.mean_squared_error(y_test,predict)
```
